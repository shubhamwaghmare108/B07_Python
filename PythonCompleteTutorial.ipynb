## 11. Tuple Packing & Unpacking

```python
# Packing
packed = 1,2,3     # (1,2,3)
# Unpacking
a,b,c = packed
print(a,b,c)
# Swapping
x,y = 5,10
x,y = y,x
print(x,y)
# Extended unpacking
first, *rest = [10,20,30,40]
print(first, rest)   # 10 [20,30,40]
```

## 12. Set Methods

## 15. Shallow vs Deep Copy

```python
import copy
original = [1,2,[3,4]]
shallow = copy.copy(original)
deep = copy.deepcopy(original)

shallow[2][0] = 999
print(original)   # [1,2,[999,4]]  (changed because inner list shared)
print(deep)       # [1,2,[3,4]]    (unchanged)
```

## 20. Functions (including advanced)

```python
# Basic function
def add(a,b): return a+b

# Default arguments
def greet(name="Guest"): print(f"Hello {name}")

# *args and **kwargs
def show(*args, **kwargs):
    print(args, kwargs)

show(1,2,3, x=10, y=20)

# Decorator example
def timer(func):
    import time
    def wrapper(*args, **kwargs):
        start = time.time()
        result = func(*args, **kwargs)
        print(f"Time: {time.time()-start:.4f}s")
        return result
    return wrapper

@timer
def slow():
    import time
    time.sleep(0.1)

slow()

# Generator using yield
def countdown(n):
    while n > 0:
        yield n
        n -= 1
for val in countdown(3):
    print(val)

# Partial function
from functools import partial
def power(base, exp): return base ** exp
square = partial(power, exp=2)
print(square(5))   # 25
```

## 21. Object-Oriented Programming (OOP)

```python
class Animal:
    species = "Unknown"  # class attribute
    def __init__(self, name):
        self.name = name
    def speak(self):
        raise NotImplementedError

class Dog(Animal):
    def speak(self):
        return f"{self.name} says Woof!"

d = Dog("Rex")
print(d.speak())

# Property decorator
class Celsius:
    def __init__(self, temp=0):
        self._temp = temp
    @property
    def temperature(self):
        return self._temp
    @temperature.setter
    def temperature(self, value):
        if value < -273.15:
            raise ValueError("Too cold")
        self._temp = value

c = Celsius(25)
print(c.temperature)

# Abstract base class
from abc import ABC, abstractmethod
class Shape(ABC):
    @abstractmethod
    def area(self): pass

class Circle(Shape):
    def __init__(self, r): self.r = r
    def area(self): return 3.14 * self.r**2

print(Circle(5).area())

# Composition vs inheritance (example)
class Engine:
    def start(self): print("Engine started")
class Car:
    def __init__(self):
        self.engine = Engine()   # composition
    def start(self):
        self.engine.start()
my_car = Car()
my_car.start()
```

## 22. Metaclasses (brief example)

```python
class SingletonMeta(type):
    _instances = {}
    def __call__(cls, *args, **kwargs):
        if cls not in cls._instances:
            cls._instances[cls] = super().__call__(*args, **kwargs)
        return cls._instances[cls]

class Database(metaclass=SingletonMeta):
    pass

db1 = Database()
db2 = Database()
print(db1 is db2)   # True
```

## 24. File Handling

```python
# Write text
with open("sample.txt", "w") as f:
    f.write("Hello\nWorld")

# Read text
with open("sample.txt", "r") as f:
    content = f.read()
print(content)

# Binary copy
with open("sample.txt", "rb") as src, open("copy.txt", "wb") as dst:
    dst.write(src.read())

# CSV example
import csv
with open("data.csv", "w", newline="") as f:
    writer = csv.writer(f)
    writer.writerow(["Name","Age"])
    writer.writerow(["Alice",30])

# JSON example
import json
with open("data.json", "w") as f:
    json.dump({"name":"Bob","age":25}, f)

# Read JSON back
with open("data.json", "r") as f:
    data = json.load(f)
print(data)
```

## 26. Regular Expressions (`re` module)

```python
import re
text = "The rain in Spain falls mainly on the plain."
match = re.search(r"Spain", text)
print(match.group() if match else None)
print(re.findall(r"\b\w+ain\b", text))   # ['rain', 'Spain', 'plain']
print(re.sub(r"ain", "***", text))

# Metacharacters
pattern = r"\d{3}-\d{2}-\d{4}"
print(re.findall(pattern, "SSN: 123-45-6789"))
```
