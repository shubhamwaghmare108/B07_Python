```python
def factorial(val):
    if val >= 0 and isinstance(val,int):
        if val == 1 or val == 0: # base condition
            return 1
        return val*factorial(val-1) #recursive function
    else:
        print("Invalied Data")
```

```python
def sum_digits(num:int)->int:
    """Takes a num and return sum of all digits
        Example:
            sum_digits(1234) ==> return 1+2+3+4= 10"""
    
    if isinstance(num,int):
        result = 0
        while num > 0:
            result += num % 10
            num //= 10
        return result
    else:
        raise ValueError(f"Invalied Data,function expect integer type of data but got {type(num)}.")
```

```python
def count_vowel(in_str):
    vowel = 'aeiou'
    count = 0
    if isinstance(in_str,str):
        in_str = in_str.lower()
        for i in in_str:
            if i in vowel:
                count += 1
        return count
    else:
        raise ValueError("Not a string data as expected")
```

```python
count_vowel("I love python")
```

```python
count_vowel("education")
```

```python
def reverseAstr(string):
    string = str(string)
    result = ""
    for i in range(-1,(len(string)*-1)-1,-1):
        result += string[i]
    return result
        
```

```python
reverseAstr('python')
```

```python
reverseAstr(123456)
```

```python
def counter(string):
    
    result = dict().fromkeys(string,0)
    
    for i in string:
        result[i] += 1
    
    lis = []
    lis1 = result.items()
    
    for i,j in lis1:
        lis.append(f"{i}:{j}")
    
    return " ".join(lis)
```

```python
counter('banana')
```

```python
def second_large(lis):
    try:
        lis.sort(reverse=True)
    except Exception as e:
        return f"error: {e}"
    else:
        return lis[1]
        
second_large([1,2,8,9,45,76,58,58.2])
```
