```python
com = 0
player = 0
```

```python
list_choice = ["rock","paper","sciser"]
```

```python
com_choice = np.random.choice(list_choice)
```

```python
player_choice = list_choice[int(input(f"{list(enumerate(list_choice))}"))]
```

```python
if com_choice == player_choice:
    print("tie")
elif com_choice == "rock" and player_choice == "sciser":
    com += 1
    print("com win")
else:
    player += 1
    print("player win")
```
