```python
current_light = input("Enter trafic light color\n").lower().strip()

if current_light == "red":
    print("Traffic Light is RED: STOP! Do not proceed.")

elif current_light == "yellow":
    print("Traffic Light is YELLOW: Prepare to stop.\
    Clear the intersection if it is safe to do so.")

elif current_light == "green":
    print("Traffic Light is GREEN: PROCEED with caution.")

else:
    print("Invalid traffic light state. Something is wrong!")
```
