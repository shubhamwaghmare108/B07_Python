```python
class AccountHolder:
    acount=1
    def __init__(self,name,dob,ac,mob,add):
        self.__acno=self.acount
        self._name=name
        self.dob=dob
        self.ac=ac
        self.mob=mob
        self.add=add
        self.bal=0
        AccountHolder.acount+=1
    def get_name(self):
        return self._name
    def set_name(self,new):
        self._name=new
        return f"sucessfully set new name"
    def get_acno(self):
        return self.__acno
       
```
