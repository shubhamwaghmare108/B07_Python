### short circuit evaluation

1. and → Logical Operator

Used for logical (boolean) operations.
Works only with True/False values (or values that can be treated as True/False).

✔ Behavior

Returns the first False-like value, otherwise the last value.

Performs short-circuit evaluation (stops early if possible).

✔ Example

```python
True and True 
```

```python
True and False 
```

2. or → Logical OR operator

Used for boolean logic.

✔ Works on:

Boolean values (True, False)

Any object with truthiness (0, None, empty list, etc.)

✔ Behavior:

Returns first True-like value, otherwise the last one.

Performs short-circuit evaluation
(stops as soon as the result is known).

✔ Examples:

```python
True or False   # True
```

```python
False or True   # True
```

```python
0 or 5     # 5  (0 is False → return 5)
```

```python
0 or 5 or 0    # 10 (10 is True → return 10)
```

```python
def test():
    print("Function executed")
    return True

1 or test()   # Does NOT call test()
```

3. not → Logical NOT operator

Used for boolean logic.

✔ Works on:

Boolean values (True, False)

Truthy / falsy values (0, None, [], {})

✔ Behavior:

Converts the value to True/False, then flips it.

✔ Examples:

```python
not True      # False
```

```python
not False     # True
```

```python
not 0         # True      (0 is False)
```

```python
not 5         # False     (5 is True)
```

```python
not []        # True
```

```python
not [1,2]     # False
```

Moves bits left → multiplies by 2ⁿ.

Python's is operator is affected by internal object caching mechanisms, which are performance optimizations implemented by the CPython interpreter. The is operator checks for object identity (same memory location), not value equality, and these caching behaviors determine whether two identical literals point to the same object in memory. 
Key aspects of object caching and the is operator:
is vs. ==:
The is operator returns True if two variables point to the exact same object in memory (same id()).
The == operator returns True if two objects have the same value, even if they are different objects in memory.
Immutable Types are Cached: Python often caches and reuses instances of certain immutable types to save memory and increase execution speed. If multiple parts of the code use the same immutable value, the interpreter might use a reference to the single existing object. This can lead to the is operator returning True for two seemingly separate creations of the same value.
Specific Caching Examples (CPython):
Integers: Integers in the range of -5 to 256 are always cached. Any variable assigned an integer in this range will reference the same object. Integers outside this range are generally not guaranteed to be cached and will usually be different objects, so 257 is 257 will likely be False, while 256 is 256 is True.
Strings: Small strings composed of identifier characters are interned (cached) at compile time. This means identical string literals defined in the same code block (like a module or function) will likely be the same object, leading to True with is.
None, True, False: These singletons are always cached and are guaranteed to be unique objects.
Mutable Types are Not Cached: Mutable objects like lists, dictionaries, and sets are not implicitly cached by the interpreter in the same way. Creating two empty lists, [] and [], will always result in two distinct objects in memory, so [] is [] will always be False.
Consequence for Your Code: You should never rely on the caching behavior when using the is operator in your general code, as it is an implementation detail of the specific Python interpreter (e.g., CPython, PyPy) and can lead to unexpected results or break if the implementation changes. Always use the == operator for value comparisons, and only use is when you explicitly need to check for object identity (e.g., checking if a variable is None). 
For application-specific caching, Python provides tools like the @lru_cache and @cached_property decorators in the functools module, which allow you to cache function results or object properties explicitly.
