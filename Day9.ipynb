## break keyword in for loop

```python
for i in range(10):
    if i==5:
        print("5 found")
        break
    else:
        print(f"not 5 but {i}")
```

```python
for i in range(10):
    print(i)
    break
print("Out")
```

```python
for i in range(10):
    print(i)
    #break
print("Out")
```

```python
for i in range(1,10):
    print(i)
    if i % 2 == 0:
        print(i)
        print("Even")
        break
```

```python
for i in  range(10):
    if i == 5:
        continue
    print(i)
    print(123)
```
