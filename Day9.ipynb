## For Loop

```python
for num in range(10): # 0-9 all included
    print(num)
```

```python
for num in range(1,11):# 1-10 all included
    print(num*2)
```

## break keyword in for loop

```python
for i in range(10):
    if i==5:
        print("5 found")
        break
    else:
        print(f"not 5 but {i}")
```

```python
for i in range(10):
    print(i)
    break
print("Out")
```

```python
for i in range(10):
    print(i)
    #break
print("Out")
```

```python
for i in range(1,10):
    if i % 2 == 0:
        print(i)
        print("Even")
        break
```

```python
for i in  range(10):
    if i == 5:
        continue
    print(i)
    print(123)
```

```python
for i in range(10):
    if i == 15:
        break
else:
    print("15 not found")
```

```python
for i in range(10):
    if i == 5:
        break
else:
    print("5 not found")
```

```python
# if break then else will not execute.
```

```python
for i in range(10):
    if i == 5:
        continue
else:
    print("5 not found")
```

```python
lis=[1,5,6,4,5]
```

```python
for i in lis:
    print(f"In with i={i}")
    if i == 5:
        print("5 found")
        break
else:
    print("5 not found")
print("Out")
```
