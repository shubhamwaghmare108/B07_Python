```python
class InvaliedAgeError(Exception):
    pass
```

```python
try:
    input_age = int(input("Enter your age: "))
    if input_age < 0:
        raise InvaliedAgeError("Age cannot be negative.")
except InvaliedAgeError as e:
    print(f"Error: {e}")
except ValueError:
    print("Error: Please enter a valid integer for age.")
```
