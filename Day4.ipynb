# Indexing And Slicing in String

```python
name = "Shubham Waghmare"
for i,j in enumerate(name):
    print(i,j)
```

```python
reversed_name = name[::-1]
i = -1
for j in reversed_name:
    print(i,j)
    i-=1
```
