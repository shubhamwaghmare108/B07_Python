# Complete Pandas Tutorial

Pandas is the most popular Python library for data manipulation and analysis. It provides two primary data structures:
- **Series**: 1D labeled array
- **DataFrame**: 2D labeled data structure (like a spreadsheet)

This tutorial covers everything from basics to advanced operations like merging, grouping, handling missing data, and visualization.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Set display options
pd.set_option('display.max_columns', None)
np.random.seed(42)
```

## 1. Series – One-Dimensional Labeled Array

```python
# From list
s1 = pd.Series([1, 2, 3, 4, 5])
print("Default index:")
print(s1)

# With custom index
s2 = pd.Series([1, 2, 3, 4], index=['A', 'B', 'C', 'D'])
print("\nCustom index:")
print(s2)

# From dictionary (keys become index)
s3 = pd.Series({'a': 10, 'b': 20, 'c': 30})
print("\nFrom dict:")
print(s3)
```

```python
# Accessing elements
print("By integer position:", s2.iat[0])        # 1
print("By label:", s2['B'])                # 2
print("Slicing by label (includes end):", s2['B':'D'])  # B 2, C 3, D 4
print("Slicing by position (excludes end):", s2[1:3])   # B 2, C 3

# Vectorized operations
print("s2 + 10:\n", s2 + 10)
print("s2.mean():", s2.mean())
print("s2 > 2:\n", s2 > 2)
```

## 10. Merging and Concatenating DataFrames

- `pd.concat()` – stack DataFrames vertically (rows) or horizontally (columns)
- `pd.merge()` – join DataFrames on keys (like SQL joins)

```python
# Create sample data
df_left = pd.DataFrame({'key': ['A', 'B', 'C'], 'value_left': [1, 2, 3]})
df_right = pd.DataFrame({'key': ['B', 'C', 'D'], 'value_right': [4, 5, 6]})
print("Left DataFrame:\n", df_left)
print("\nRight DataFrame:\n", df_right)
```

```python
# Concatenate vertically (rows)
df_concat_rows = pd.concat([df_left, df_right], ignore_index=True)
print("\nConcatenated rows:\n", df_concat_rows)
```

```python
# Concatenate horizontally (columns)
df_concat_cols = pd.concat([df_left, df_right], axis=1)
print("\nConcatenated columns:\n", df_concat_cols)
```

```python
# Create sample data
df_left = pd.DataFrame({'key': ['A', 'B', 'C'], 'value_left': [1, 2, 3]})
df_right = pd.DataFrame({'key': ['B', 'C', 'D'], 'value_right': [4, 5, 6]})
print("Left DataFrame:\n", df_left)
print("\nRight DataFrame:\n", df_right)

# Concatenate vertically (rows)
df_concat_rows = pd.concat([df_left, df_right], ignore_index=True)
print("\nConcatenated rows:\n", df_concat_rows)

# Concatenate horizontally (columns)
df_concat_cols = pd.concat([df_left, df_right], axis=1)
print("\nConcatenated columns:\n", df_concat_cols)
```
