# NumPy: Array Manipulation, Mathematical & Statistical Operations, and More

This notebook covers essential NumPy topics beyond basic creation and slicing:
- Array Manipulation (reshape, resize, flatten, ravel, transpose)
- Mathematical Operations (basic arithmetic, universal functions, dot product)
- Statistical Functions (mean, median, std, var, correlation, etc.)

```python
import numpy as np

# Set seed for reproducibility
np.random.seed(42)
```

## 1. Array Manipulation

Changing the shape, size, or layout of arrays without (or with) copying data.

### `reshape()` – Change shape without changing data

Returns a **view** (if possible) of the array with a new shape. The total number of elements must remain the same.

```python
arr = np.arange(12)
print("Original:", arr)

# Reshape to 3x4
reshaped = arr.reshape(3, 4)
print("Reshaped (3x4):\n", reshaped)

# Using -1 for automatic dimension
auto = arr.reshape(2, -1)  # 2 rows, columns inferred as 6
print("Auto reshape (2x6):\n", auto)

# Check if it's a view (shares memory)
print("Shares memory?", np.shares_memory(arr, reshaped))  # True
```

### `resize()` – Change shape and size, may fill with zeros

- `ndarray.resize()`: in‑place, can change total size (fills with zeros if larger, truncates if smaller).
- `np.resize()`: returns a new array, repeats values if needed.

```python
# In-place resize (modifies original)
a = np.array([1, 2, 3, 4])
a.resize(2, 3)  # Now 2 rows, 3 columns (fills extra with zeros)
print("In-place resize:\n", a)

# np.resize (returns new array, repeats values)
b = np.array([1, 2, 3])
c = np.resize(b, (2, 4))
print("np.resize (repeating):\n", c)
```

### `flatten()` vs `ravel()` – Flatten to 1D

- `flatten()`: always returns a **copy**.
- `ravel()`: returns a **view** whenever possible (more memory efficient).

```python
mat = np.array([[1, 2], [3, 4]])

flat_copy = mat.flatten()
flat_view = mat.ravel()

print("Flatten (copy):", flat_copy)
print("Ravel (view):", flat_view)

# Modifying the view modifies original
flat_view[0] = 99
print("Original after modifying ravel view:\n", mat)

# Flatten copy does not affect original (reset)
mat = np.array([[1, 2], [3, 4]])  # reinitialize
flat_copy[0] = 88
print("Original after flatten copy modification:\n", mat)
```

### `transpose()` and `.T` – Swap axes

Returns a view.

```python
mat = np.arange(6).reshape(2, 3)
print("Original:\n", mat)
print("Transpose (.T):\n", mat.T)
print("transpose(1,0):\n", mat.transpose(1, 0))

# For 3D arrays
cube = np.arange(24).reshape(2, 3, 4)
print("Original shape:", cube.shape)
print("Transpose (0,2,1) shape:", cube.transpose(0, 2, 1).shape)
```

## 2. Mathematical Operations

NumPy provides fast, vectorized operations for element-wise math and linear algebra.

### Basic Arithmetic (Element‑wise)

```python
A = np.array([1, 2, 3])
B = np.array([4, 5, 6])

print("A + B =", A + B)
print("A - B =", A - B)
print("A * B =", A * B)
print("A / B =", A / B)
print("A ** 2 =", A ** 2)
print("A % 2 =", A % 2)
```

### Universal Functions (ufuncs)

Fast element‑wise mathematical functions.

```python
x = np.linspace(0, np.pi, 4)
print("x =", x)
print("sin(x) =", np.sin(x))
print("cos(x) =", np.cos(x))
print("exp(x) =", np.exp(x))
print("log(x) =", np.log(x))  # natural log
print("sqrt(x) =", np.sqrt(x))
print("power(x, 2) =", np.power(x, 2))
```

### Dot Product & Matrix Multiplication

- `np.dot()` or `@` operator
- `np.matmul()`

```python
v = np.array([1, 2])
w = np.array([3, 4])
print("Dot product (v·w):", np.dot(v, w), "or", v @ w)

M = np.array([[1, 2], [3, 4]])
N = np.array([[5, 6], [7, 8]])
print("Matrix multiplication:\n", M @ N)
```

### Aggregations

`sum()`, `mean()`, `min()`, `max()`, `argmin()`, `argmax()`, `cumsum()`, `cumprod()`.

```python
arr = np.array([[1, 2], [3, 4]])

print("Sum:", arr.sum())                # total sum
print("Sum axis=0 (columns):", arr.sum(axis=0))
print("Sum axis=1 (rows):", arr.sum(axis=1))
print("Mean:", arr.mean())
print("Min:", arr.min())
print("Max:", arr.max())
print("Argmin (flattened index):", arr.argmin())
print("Cumulative sum:", arr.cumsum())
print("Cumulative product:", arr.cumprod())
```

## 3. Statistical Functions

NumPy includes many descriptive statistics.

```python
data = np.array([10, 12, 23, 23, 16, 23, 21, 16])

print("Mean:", np.mean(data))
print("Median:", np.median(data))
print("Standard deviation:", np.std(data))
print("Variance:", np.var(data))
print("Percentile (50th):", np.percentile(data, 50))
print("Percentile (90th):", np.percentile(data, 90))

# Mode (most frequent value) – using unique
values, counts = np.unique(data, return_counts=True)
mode = values[np.argmax(counts)]
print("Mode:", mode)

# Correlation coefficient
x = np.array([1, 2, 3, 4])
y = np.array([2, 4, 6, 8])
print("Correlation matrix:\n", np.corrcoef(x, y))

# Covariance
print("Covariance matrix:\n", np.cov(x, y))

# Histogram
hist, bin_edges = np.histogram(data, bins=5)
print("Histogram counts:", hist)
print("Bin edges:", bin_edges)
```

## 4. Array Operations (Indexing, Slicing, Masking, Fancy, `where`, Set Operations)

### Basic Indexing & Slicing (Recap)

```python
arr = np.arange(10)
print("arr[2:7:2] =", arr[2:7:2])

mat = np.arange(9).reshape(3, 3)
print("mat[1, 2] =", mat[1, 2])
print("mat[:2, 1:] =", mat[:2, 1:])
```

### Boolean Indexing (Masking)

Select elements that satisfy a condition.

```python
arr = np.array([1, 2, 3, 4, 5])
mask = arr > 2
print("Mask:", mask)
print("arr[mask]:", arr[mask])

# Inline condition
print("arr[arr % 2 == 0]:", arr[arr % 2 == 0])

# Modifying with mask
arr[arr > 3] = 99
print("After modification:", arr)
```

### Fancy Indexing – Using integer arrays

```python
arr = np.array([10, 20, 30, 40, 50])
indices = [0, 2, 4]
print("arr[indices]:", arr[indices])

# 2D fancy indexing
mat = np.arange(12).reshape(3, 4)
rows = np.array([0, 2])
cols = np.array([1, 3])
print("mat[rows, cols]:", mat[rows, cols])  # (0,1) and (2,3)

# Using np.ix_ for outer indexing
print("Outer indexing:\n", mat[np.ix_(rows, cols)])
```

### `np.where()` – Conditional selection

`where(condition, x, y)` returns `x` where condition is True, else `y`.

```python
arr = np.array([1, 2, 3, 4, 5])
result = np.where(arr > 3, "high", "low")
print("where condition:", result)

# Only indices where condition is True
indices = np.where(arr > 3)
print("Indices where >3:", indices)
```

### Set Operations

`unique`, `intersect1d`, `union1d`, `setdiff1d`, `in1d`.

```python
a = np.array([1, 2, 3, 2, 4])
b = np.array([3, 4, 5, 6])

print("Unique a:", np.unique(a))
print("Intersection:", np.intersect1d(a, b))
print("Union:", np.union1d(a, b))
print("Difference (a - b):", np.setdiff1d(a, b))
print("Is element of a in b?", np.in1d(a, b))
```

## 5. Logical Operations

NumPy provides both comparison operators and logical functions.

```python
a = np.array([True, False, True])
b = np.array([False, False, True])

print("a and b (logical_and):", np.logical_and(a, b))
print("a or b (logical_or):", np.logical_or(a, b))
print("not a (logical_not):", np.logical_not(a))
print("xor (logical_xor):", np.logical_xor(a, b))

# all() and any()
print("all(a):", np.all(a))
print("any(a):", np.any(a))

# Comparisons
x = np.array([1, 2, 3])
y = np.array([1, 4, 2])
print("x == y:", x == y)
print("x < y:", x < y)
print("x >= y:", x >= y)
```

## 6. String Operations (`np.char`)

Vectorized string methods for arrays of dtype `str` or `unicode`.

```python
names = np.array(["alice", "BOB", "Charlie"])

print("Upper:", np.char.upper(names))
print("Lower:", np.char.lower(names))
print("Capitalize:", np.char.capitalize(names))
print("Length:", np.char.str_len(names))

# Concatenation
greeting = np.char.add("Hello, ", names)
print("Concatenated:", greeting)

# Splitting
sentence = np.array(["one two three", "four five six"])
split = np.char.split(sentence)
print("Split:", split)

# Replace
print("Replace 'a' with 'X':", np.char.replace(names, 'a', 'X'))

# Strip whitespace
dirty = np.array(["  hello  ", "  world  "])
print("Strip:", np.char.strip(dirty))
```

## 7. Sorting and Searching

- `np.sort()` returns a sorted copy.
- `ndarray.sort()` sorts in‑place.
- `argsort()` returns the indices that would sort the array.
- `searchsorted()` finds insertion positions.
- `lexsort()` sorts by multiple keys.
- `partition()` performs a partial sort.

```python
arr = np.array([3, 1, 2, 5, 4])

# Copy sort
sorted_copy = np.sort(arr)
print("Sorted copy:", sorted_copy)
print("Original unchanged:", arr)

# In-place sort
arr.sort()
print("In-place sorted:", arr)

# argsort
arr2 = np.array([40, 10, 30, 20])
indices = np.argsort(arr2)
print("Indices that would sort:", indices)
print("arr2[indices]:", arr2[indices])

# searchsorted (array must be sorted)
sorted_arr = np.array([1, 3, 5, 7, 9])
positions = np.searchsorted(sorted_arr, [2, 5, 10])
print("Insertion positions:", positions)

# lexsort – sort by last name, then first name
last_names = np.array(['Smith', 'Jones', 'Smith', 'Adams'])
first_names = np.array(['John', 'Bob', 'Alice', 'Tom'])
ind = np.lexsort((first_names, last_names))
print("Sorted by last, then first:", list(zip(last_names[ind], first_names[ind])))

# partition – smallest k elements (not fully sorted)
arr3 = np.array([7, 2, 5, 1, 9, 3])
partitioned = np.partition(arr3, 3)  # first 3 elements are the 3 smallest (order unspecified)
print("Partitioned (k=3):", partitioned)
```
