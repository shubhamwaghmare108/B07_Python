```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("Titanic.xlsx")
```

```python
df
```

```python
df.set_index("PassengerId",inplace = True)
```

```python
df
```

```python
survival_rate_by_sex = df.groupby('Sex')['Survived'].sum()
```

```python
survival_rate_by_sex
```

```python
survival_rate_by_sex.plot(kind='bar', color=['skyblue', 'salmon'])
plt.title('Survival Rate on the Titanic by Gender')
plt.xlabel('Gender')
plt.ylabel('Survival Rate')
plt.xticks(rotation=90)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
```

```python
df.groupby("Pclass")["Survived"].sum().plot(kind="barh")
plt.xticks(rotation=0)
```

```python
x = np.linspace(0, 10, 100)  # 100 points between 0 and 10
y = np.sin(x)

plt.plot(x, y, label="sin curve", color="blue", linestyle="--", marker="x")
plt.title("Basic Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.grid(False)
plt.show()
```

```python
y2 = np.tan(x)

plt.plot(x, y, label="sin(x)", color="blue")
plt.plot(x, y2, label="tan(x)", color="red")
plt.title("Multiple Line Plots")
plt.xlabel("X")
plt.ylabel("Value")
plt.legend()
plt.show()
```

```python
fig, axs = plt.subplots(2, 3, figsize=(15,5)) #oo

axs[0,0].plot(x, y, color="blue"); axs[0,0].set_title("sin(x)")
axs[0,1].plot(x, y2, color="red"); axs[0,1].set_title("tan(x)")
axs[1,0].scatter(x, y, color="green"); axs[1,0].set_title("Scatter")
axs[1,2].hist(y, bins=20, color="purple"); axs[1,1].set_title("Histogram")

plt.tight_layout()
plt.show()
```

```python

categories = ["A", "B", "C", "D"]
values = [23, 45, 56, 78]

fig, ax = plt.subplots()
bars = ax.bar(categories, values, color="orange")

# Add value labels
ax.bar_label(bars, label_type="edge")

plt.title("Bar Chart with Data Labels")
plt.show()
```

```python
x = np.linspace(0, 10, 100)

plt.subplot(2, 2, 1)
plt.plot(x, np.sin(x))
plt.title("Sine Wave")

plt.subplot(2, 2, 2)
plt.plot(x, np.cos(x))
plt.title("Cosine Wave")

plt.subplot(2, 2, 3)
plt.plot(x, np.tan(x))
plt.title("Tangent")

plt.subplot(2, 2, 4)
plt.hist(np.random.randn(1000), bins=30)
plt.title("Histogram")

plt.tight_layout()
plt.show()
```

| String      | Short Code | Appearance    |

| `'dashed'`  | `'--'`     | – – – – – – – |
| `'dotted'`  | `':'`      | ···········   |

```python
fig, axs = plt.subplots(3, 1, sharex=False, figsize=(6, 8))

x = np.linspace(0, 10, 100)

axs[0].plot(x, np.sin(x))
axs[0].set_title("sin(x)")

axs[1].plot(x, np.cos(x), color="red")
axs[1].set_title("cos(x)")

axs[2].plot(x, np.sin(x) + np.cos(x), color="green")
axs[2].set_title("sin(x)+cos(x)")
plt.savefig("demo.pdf",dpi=500)
plt.tight_layout()
plt.show()
```
