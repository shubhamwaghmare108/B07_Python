# Complete Matplotlib Tutorial

Matplotlib is the most widely used Python library for data visualization. It provides:
- **Fine-grained control** over every plot element
- **Publication-quality** figures
- **Integration** with NumPy, Pandas, and Jupyter
- **Extensive customization** (colors, fonts, axes, annotations)
- **Backend support** (interactive, static, web)

This tutorial covers everything from basic plots to advanced 3D visualizations.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl

# For 3D plots
from mpl_toolkits.mplot3d import Axes3D

# Set style and defaults
plt.style.use('seaborn-v0_8-darkgrid')
mpl.rcParams['figure.figsize'] = (10, 6)
mpl.rcParams['font.size'] = 12
mpl.rcParams['lines.linewidth'] = 2

# For Jupyter notebook inline display
%matplotlib inline

np.random.seed(42)
```

## 1. Basic Plotting

The core of Matplotlib is the `pyplot` module, which provides a MATLAB-like interface.

```python
# Simple line plot
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y)
plt.title('Sine Wave')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.grid(True,alpha = 0.6)
plt.show()
```

```python
plt.legend?
```

```python
# Multiple lines on same plot
y2 = np.cos(x)

plt.plot(x, y, label='sin(x)', color='blue', linestyle='-', linewidth=2)
plt.plot(x, y2, label='cos(x)', color='red', linestyle='--', linewidth=2)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Sine and Cosine')
plt.legend(loc=3)
plt.grid(True, alpha=0.5)
plt.show()
```

## 2. Subplots – Multiple Axes in One Figure

Use `plt.subplots()` to create a grid of subplots.

```python
# 2x2 subplots
fig, axes = plt.subplots(2, 2, figsize=(12, 8))

# First subplot
axes[0, 0].plot(x, np.sin(x), 'b-')
axes[0, 0].set_title('Sine')
axes[0, 0].set_xlabel('x')
axes[0, 0].set_ylabel('sin(x)')

# Second subplot
axes[0, 1].plot(x, np.cos(x), 'r--')
axes[0, 1].set_title('Cosine')
axes[0, 1].set_xlabel('x')
axes[0, 1].set_ylabel('cos(x)')

# Third subplot
axes[1, 0].plot(x, np.tanh(x), 'g-.')
axes[1, 0].set_title('Tanh')
axes[1, 0].set_xlabel('x')
axes[1, 0].set_ylabel('tanh(x)')

# Fourth subplot
axes[1, 1].scatter(x[::5], np.sin(x[::5]), c='purple', alpha=0.6)
axes[1, 1].set_title('Scatter')
axes[1, 1].set_xlabel('x')
axes[1, 1].set_ylabel('sin(x)')

plt.tight_layout()
plt.show()
```

## 3. Common Plot Types

### Line Plot
Already covered.

### Scatter Plot

```python
# Scatter plot with color mapping
n = 200
x_scatter = np.random.randn(n)
y_scatter = np.random.randn(n)
colors = np.random.rand(n)
sizes = np.random.randint(20, 200, n)

plt.scatter(x_scatter, y_scatter, c=colors, s=sizes, alpha=0.6, cmap='viridis')
plt.colorbar(label='Color value')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Scatter Plot with Variable Size and Color')
plt.show()
```

```python
# Bar plot
categories = ['A', 'B', 'C', 'D', 'E']
values = [23, 45, 56, 78, 34]
plt.figure(figsize=(5,3))
plt.bar(categories, values, color='skyblue', edgecolor='navy', alpha=0.7)
plt.xlabel('Category')
plt.ylabel('Value')
plt.title('Bar Chart')
plt.show()
```

```python
# Horizontal bar chart
plt.figure(figsize=(5,3))
plt.barh(categories, values, color='lightcoral')
plt.xlabel('Value')
plt.ylabel('Category')
plt.title('Horizontal Bar Chart')
plt.show()
```

```python
# Histogram
data = np.random.randn(1000)

plt.hist(data, bins=30, color='green', alpha=0.7, edgecolor='black')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram')
plt.axvline(data.mean(), color='red', linestyle='--', label=f'Mean: {data.mean():.2f}')
plt.legend()
plt.show()
```

```python
# Box plot (shows median, quartiles, outliers)
np.random.seed(45)
data_by_group = [np.random.normal(0, 1, 100) for _ in range(4)]

plt.boxplot(data_by_group, label=['Group 1', 'Group 2', 'Group 3', 'Group 4'], patch_artist=True)
plt.ylabel('Value')
plt.title('Box Plot')
plt.grid(axis='y', alpha=0.3)
plt.show()
```

```python
# Pie chart
sizes = [25, 30, 20, 15, 10]
labels = ['Category A', 'Category B', 'Category C', 'Category D', 'Category E']
colors = ['gold', 'lightblue', 'lightgreen', 'coral', 'purple']
plt.figure(figsize=(6,3))
plt.pie(sizes, labels=labels, colors=colors, autopct='%0.3f%%', startangle=0, explode=(0.08, 0.03, 0, 0, 0))
plt.title('Pie Chart')
plt.axis('equal')
plt.show()
```

## 4. Customization

Matplotlib gives you complete control over every visual element.

```python
plt.xlabel?
```

```python
# Customizing lines, markers, and colors
x = np.linspace(0, 5, 20)
y1 = x
y2 = x**2
y3 = np.sqrt(x)

plt.plot(x, y1, 'ro-', label='Linear', linewidth=2, markersize=8)
plt.plot(x, y2, 'gs--', label='Quadratic', linewidth=3, markersize=6)
plt.plot(x, y3, 'b<-.', label='Square root', linewidth=1, markersize=3)

plt.xlabel('X', fontsize=14, fontweight='bold',loc ='right')
plt.ylabel('Y', fontsize=14, fontweight='bold')
plt.xticks(fontsize=14,color='red')
plt.title('Customized Line Plot', fontsize=16, fontweight='bold',color = 'blue')
plt.legend(loc='upper left', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
# Adding annotations
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y, 'b-')

# Annotate the maximum point
max_idx = np.argmax(y)
plt.plot(x[max_idx], y[max_idx], 'ro', markersize=10)
plt.annotate(f'Max: {y[max_idx]:.2f}', xy=(x[max_idx], y[max_idx]), xytext=(x[max_idx]+1, y[max_idx]-0.2),
             arrowprops=dict(facecolor='black', shrink=0.05, width=1, headwidth=8))

# Add text box
plt.text(0, 0, 'Sine wave\nPeriodic function', bbox=dict(facecolor='yellow', alpha=0.5))

plt.title('Line Plot with Annotations')
plt.show()
```

## 5. Colormaps and Color Bars

Use colormaps to represent a third dimension (e.g., density, time, value).

```python
# 2D color plot (pcolormesh / imshow)
X, Y = np.meshgrid(np.linspace(-3, 3, 100), np.linspace(-3, 3, 100))
Z = np.sin(X) * np.cos(Y)

plt.figure(figsize=(8, 6))
plt.pcolormesh(X, Y, Z, cmap='plasma', shading='auto')
plt.colorbar(label='sin(x) * cos(y)')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('2D Colormap Plot')
plt.show()
```

## 6. 3D Plots

Matplotlib supports 3D plotting via the `mplot3d` toolkit.

```python
# 3D scatter plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

n_pts = 200
xs = np.random.randn(n_pts)
ys = np.random.randn(n_pts)
zs = np.random.randn(n_pts)

ax.scatter(xs, ys, zs, c=zs, cmap='viridis', s=30, alpha=0.6)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('3D Scatter Plot')
plt.show()
```

```python
# 3D surface plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

X, Y = np.meshgrid(np.linspace(-5, 5, 50), np.linspace(-5, 5, 50))
Z = np.sin(np.sqrt(X**2 + Y**2))

surf = ax.plot_surface(X, Y, Z, cmap='coolwarm', linewidth=0, antialiased=True)
fig.colorbar(surf, shrink=0.5, aspect=10, label='sin(r)')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('3D Surface Plot')
plt.show()
```

```python
# 3D contour plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

ax.contour3D(X, Y, Z, 50, cmap='viridis')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('3D Contour Plot')
plt.show()
```

## 7. Plotting with Pandas

Pandas DataFrames have built‑in plotting methods that use Matplotlib.

```python
# Create sample DataFrame
dates = pd.date_range('2024-01-01', periods=100, freq='D')
df_plot = pd.DataFrame({
    'sales': np.random.randn(100).cumsum() + 100,
    'expenses': np.random.randn(100).cumsum() + 90,
    'profit': np.random.randn(100).cumsum() + 10
}, index=dates)

# Line plot
df_plot.plot(title='Financial Trends', figsize=(10, 6))
plt.xlabel('Date')
plt.ylabel('Amount')
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# Histogram from DataFrame
df_plot.hist(bins=30, alpha=0.7, figsize=(10, 6))
plt.suptitle('Histograms of Financial Metrics', y=1.02)
plt.tight_layout()
plt.show()
```

```python
# Scatter matrix (pair plot)
from pandas.plotting import scatter_matrix

df_pair = pd.DataFrame(np.random.randn(500, 3), columns=['A', 'B', 'C'])
scatter_matrix(df_pair, alpha=0.5, figsize=(10, 10), diagonal='hist')
plt.suptitle('Scatter Matrix', y=1.02)
plt.show()
```

## 8. Saving Figures

Save plots to various formats: PNG, PDF, SVG, EPS, etc.

```python
plt.figure(figsize=(8, 6))
plt.plot(x, np.sin(x))
plt.title('Sine Wave')
plt.xlabel('x')
plt.ylabel('sin(x)')

# Save as PNG with high DPI
plt.savefig('sine_wave.png', dpi=300, bbox_inches='tight')
print("Saved as 'sine_wave.png'")

# Save as PDF (vector format)
plt.savefig('sine_wave.pdf', bbox_inches='tight')
print("Saved as 'sine_wave.pdf'")

# Save as SVG (editable in Illustrator/Inkscape)
plt.savefig('sine_wave.svg', bbox_inches='tight')
print("Saved as 'sine_wave.svg'")

plt.show()
```

## 9. Matplotlib Styles

Apply predefined styles or create your own.

```python
# List available styles
print("Available styles:", plt.style.available[:10])

# Demo different styles
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

styles = ['default', 'seaborn-v0_8-darkgrid', 'ggplot', 'fivethirtyeight']
for ax, style in zip(axes.flatten(), styles):
    with plt.style.context(style):
        ax.plot(x, np.sin(x), label='sin(x)')
        ax.plot(x, np.cos(x), label='cos(x)')
        ax.set_title(f'Style: {style}')
        ax.legend()
        ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

## 10. Advanced Plot Types

### Violin Plot (combines box plot and kernel density)

```python
data_violin = [np.random.normal(0, std, 100) for std in range(1, 5)]

plt.violinplot(data_violin, showmeans=True, showmedians=True)
plt.xticks(range(1, 5), ['Std=1', 'Std=2', 'Std=3', 'Std=4'])
plt.ylabel('Value')
plt.title('Violin Plot')
plt.grid(axis='y', alpha=0.3)
plt.show()
```

### Heatmap

```python
# Create correlation matrix
df_corr = pd.DataFrame(np.random.randn(10, 10), columns=[f'Var{i}' for i in range(1, 11)])
corr = df_corr.corr()

plt.figure(figsize=(10, 8))
plt.imshow(corr, cmap='coolwarm', interpolation='nearest')
plt.colorbar(label='Correlation')
plt.xticks(range(10), corr.columns, rotation=45)
plt.yticks(range(10), corr.columns)
plt.title('Correlation Heatmap')

# Add text annotations
for i in range(10):
    for j in range(10):
        plt.text(j, i, f'{corr.iloc[i, j]:.2f}', ha='center', va='center', color='black')

plt.tight_layout()
plt.show()
```

### Error Bars

```python
x = np.arange(5)
y = [10, 15, 20, 25, 30]
y_err = [2, 3, 2.5, 4, 3]

plt.errorbar(x, y, yerr=y_err, fmt='o-', capsize=5, capthick=2, elinewidth=2, markersize=8)
plt.xlabel('Category')
plt.ylabel('Value')
plt.title('Line Plot with Error Bars')
plt.xticks(x, ['A', 'B', 'C', 'D', 'E'])
plt.grid(alpha=0.3)
plt.show()
```

### Inset Axes (Zoomed View)

```python
fig, ax = plt.subplots(figsize=(8, 6))

x = np.linspace(0, 10, 1000)
y = np.sin(x) * np.exp(-0.1 * x)
ax.plot(x, y)
ax.set_title('Main Plot')

# Inset axes
axins = ax.inset_axes([0.6, 0.6, 0.3, 0.3])
axins.plot(x, y)
axins.set_xlim(2, 4)
axins.set_ylim(-0.5, 0.5)
axins.set_title('Zoomed')
ax.indicate_inset_zoom(axins)

plt.show()
```

## 11. Interactive Features (Zoom, Pan, Widgets)

Use `%matplotlib notebook` or `mplcursors` for interactivity. Here’s a simple slider example.

```python
from matplotlib.widgets import Slider

fig, ax = plt.subplots()
plt.subplots_adjust(bottom=0.25)

t = np.linspace(0, 2*np.pi, 500)
initial_freq = 1
line, = ax.plot(t, np.sin(initial_freq * t), lw=2)
ax.set_xlabel('t')
ax.set_ylabel('Amplitude')
ax.set_title('Interactive Sine Wave')
ax.grid(True, alpha=0.3)

ax_slider = plt.axes([0.25, 0.1, 0.65, 0.03])
slider = Slider(ax_slider, 'Frequency', 0.1, 10.0, valinit=initial_freq)

def update(val):
    freq = slider.val
    line.set_ydata(np.sin(freq * t))
    fig.canvas.draw_idle()

slider.on_changed(update)
plt.show()
```

## 12. Performance Tips for Large Datasets

- Use `rasterized=True` for dense scatter plots
- Use `linewidth=0` for scatter plots with many points
- Use `step` instead of `plot` for large step functions
- Downsample data before plotting
- Use `hexbin` for 2D density heatmaps

```python
# Hexbin plot (efficient for large datasets)
x_hex = np.random.randn(10000)
y_hex = x_hex * 0.5 + np.random.randn(10000) * 0.2

plt.hexbin(x_hex, y_hex, gridsize=50, cmap='YlOrRd', mincnt=1)
plt.colorbar(label='Count in bin')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Hexbin Plot (Density Heatmap)')
plt.show()
```
