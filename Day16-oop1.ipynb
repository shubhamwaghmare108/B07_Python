```python
class complexx:
    """
    we want to built an implimentation of python complex class.
    """
    __counter = 0
    def __init__(self,real,imag): # constructur method-- initilizer #__dgfh__ dunder method
        self.__real = float(real) # private
        self.__imag = float(imag)
        self._iscomplex = True
        complexx.__counter+=1
    
    def __repr__(self):
        if '-' in str(self.__imag):
            return  f"({self.__real}{self.__imag}i)"
        return f"({self.__real}+{self.__imag}i)"

    def __add__(self,other):
        real = self.__real + other.__real
        imag = self.__imag + other.__imag
        
        return complexx(real,imag)
    
    def __sub__(self,other):
        real = self.__real - other.__real
        imag = self.__imag - other.__imag
        
        return complexx(real,imag)
    def __mul__(self,other):
        real = self.__real * other.__real
        imag = self.__imag * other.__imag
        
        return complexx(real,imag)
    
    def __eq__(self,other):
        return (self.__real == other.__real) and (self.__imag == other.__imag)

    def __gt__(self,other):
        return (self.__real +self.__imag) > (other.__real + other.__imag)

    # getter method
    @property
    def real(self,passwd):
        if passwd == '123':
            return self.__real
        else:
            print("Bhag yahase")

    # setter method
    @real.setter
    def real(self):
        passwd = input("Enter Passwd:\n")
        if passwd == '123':
            new = input("Enter new value:\n")
            self.__real =float(new)
            print("Value updated sucessfully")
        else:
            print("Invalied Credential,Try Again....")

    @classmethod
    def get_counter(cls):
        return cls.__counter
```
