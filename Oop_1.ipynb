```python
class mission:
    def __init__(self,mission_code,mission_name,mission_location,mission_aim,mission_status):
        self.mission_code = mission_code
        self.mission_name = mission_name
        self.mission_location = mission_location
        self.mission_aim = mission_aim
        self.mission_status = mission_status    

    def __str__(self):
        return f"Mission Code: {self.mission_code}, Mission Name: {self.mission_name}, Location: {self.mission_location}, Aim: {self.mission_aim}, Status: {self.mission_status}"

    def transfer_mission_records(self, from_spy, to_spy):
        from_spy.delete_mission(self.mission_code)
        to_spy.add_mission(self)
```

```python
class Spy:
    """Creating a spy class"""
    def __init__(self,code,password,name,real_name,age,nationality,
                 real_nationality,occupation,real_occupation,address,real_address,mission_count,status='active'):
        self.__code = code
        self.__password = password
        self.name = name
        self.__name = real_name
        self.age = age
        self.nationality = nationality
        self.__nationality = real_nationality
        self.occupation = occupation
        self.__occupation = real_occupation
        self.address = address
        self.__address = real_address
        self.__mission_count = mission_count
        self.__mission_history = []
        self.__status = status

    def Identity(self,code = None,password = None):
        if code == self.__code and password == self.__password:
            return {'name':self.__name,'nationality' : self.__nationality,'age':self.age,
                    'occupation':self.__occupation,
                  'address': self.__address,'mission_count' :self.__mission_count,
                   'mission_history': self.__mission_history,'status': self.__status}
        else:
            return {'name':self.name,'nationality' : self.nationality,
                    'occupation':self.occupation,'address': self.address}
    def add_mission(self,mission):
        self.__mission_history.append(mission)
        self.__mission_count += 1 
    def get_mission_history(self):
        return self.__mission_history
    def get_mission_count(self):
        return self.__mission_count
    def delete_mission(self,mission_code):
        for mission in self.__mission_history:
            if mission.mission_code == mission_code:
                self.__mission_history.remove(mission)
                self.__mission_count -= 1
                return True
        return False

    def transfer_mission(self,mission_code,other_spy):
        for mission in self.__mission_history:
            if mission.mission_code == mission_code:
                self.__mission_history.remove(mission)
                self.__mission_count -= 1
                other_spy.add_mission(mission)
                mission.transfer_mission_records(other_spy)
                return True
        return False
```

```python
class point:
    """Implementing 2D point class"""
    __origin = (0,0)
    def __init__(self,x,y):
        self.__x = x
        self.__y = y
    def __str__(self):
        return f"Point({self.__x},{self.__y})"
    def __repr__(self):
        return f"({self.__x},{self.__y})"
    #polymorphism
    def distance(self,other_point=point(0,0),distance_type='euclidean'):
        if distance_type == 'euclidean':
            return ((self.__x - other_point.__x) ** 2 + (self.__y - other_point.__y) ** 2) ** 0.5
        elif distance_type == 'manhattan':
            return abs(self.__x - other_point.__x) + abs(self.__y - other_point.__y)

    def move_x(self,dx):
        self.__x += dx

    def move_y(self,dy):
        self.__y += dy

    def move_to_origin(self):
        self.__x = 0
        self.__y = 0

    def change_coordinates(self,new_x,new_y):
        self.__x = new_x
        self.__y = new_y

    def origin(self):
        return self.get_origin()

    @classmethod
    def get_origin(cls):
        return cls.__origin   
    @property
    def point(self):
        return (self.__x,self.__y)
    @point.setter
    def point(self,new_point: tuple):
        self.__x = new_point[0]
        self.__y = new_point[1]

    @point.deleter
    def point(self):
        del self.__x
        del self.__y
```
