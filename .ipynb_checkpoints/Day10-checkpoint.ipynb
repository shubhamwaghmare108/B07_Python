# While loop in python

```python
num = 10
while num ==10:
    print(f"Ok for num = {num}")
    num += 3
print("Out")
```

```python
num = 100
while num <10:
    print("In")
    print(f"Ok for num = {num}")
    num += 1
print("Out")
```

```python
lis = [1,5,65,89]
```

```python
i = 0
while lis[i] !=10:
    print(f"Not 10 but {lis[i]}")
    i +=1
else:
    print("10 found")
print("Out")
```

```python
lis = [1,5,65,10,89]
i = 0
while lis[i] !=10:
    print(f"Not 10 but {lis[i]}")
    i +=1
else:
    print(f"10 found at index num= {i}")
print("Out")
```

```python
# infinite loop
lis = [1,5,65,10,89]
i = 0
while lis[i] !=10:
    pass
else:
    print("loop ended")
print("Out")
```

```python
lis = [1,5,65,10,89]
i = 0
while lis[i] !=10:
    i+=1
else:
    print("loop ended")
print("Out")
```

```python
lis = [1,5,65,10,89]
i = 0
while lis[i] !=10:
    i+=1
    break
else:
    print("loop ended")
print("Out")
```

```python
lis = [1,5,65,89,10,59]
i = 0
while lis[i] !=10:
    print(lis[i])
    i+=1
    
    if i ==4:
        break
else:
    print("loop ended")
print("Out")
```

```python
lis = [1,5,65,89,10,59]
i = 0
while lis[i] !=10:
    print(lis[i])
    i+=1
    
    if i ==4:
        continue
    print(lis[i]) 
else:
    print("loop ended")
print("Out")
```

```python
lis = [1,5,6,8,6,4,2,6]
i = 0
even=[]
odd=[]
while True:
    if lis[i] % 2 == 0:
        even.append(lis[i])
        continue
    odd.append(lis[i])
    i += 1
    if len(lis)-1 == i:
        print("Index reached its limit")
        break
    
```

```python
lis = [1,5,6,8,6,"sdhg",2,6]
i = 0
even=[]
odd=[]
while len(lis)>i:
    if isinstance(lis[i],int):
        if lis[i] % 2 == 0:
            even.append(lis[i])
            i += 1
            continue
        odd.append(lis[i])
        i += 1
    else:
        i +=1
```
