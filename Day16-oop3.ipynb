```python
class Point:
    ObjectCount=0
    def __init__(self,x,y):
        self.x=x
        self.y=y
        Point.ObjectCount += 1
    
    def shift_x(self,shift):
        self.x=self.x+shift
        return (self.x,self.y)

    def shift_y(self,shift):
        self.y=self.y+shift
        return (self.x,self.y)

    def orign(self):
        self.x=0
        self.y=0
    def __repr__(self):
        return f'{{"x":{self.x},"y":{self.y}}}'

        
```

```python
class complexx:

    def __init__(self,real,imag):
        self.real=real
        self.imag=imag

    def __repr__(self):
        if "-" in str(self.imag):
            return f"{self.real} {self.imag}j"
            
        return f"{self.real} +{self.imag}j"
        

    
    def real(self):  # getter
        return self.real
    
    
    def set_real(self,new): #setter
        self.real=new
        return self
    
    def set_imag(self,new): #setter
        self.imag=new
        return self
```
