```python
def minus(x,y,/,z,c,*,d,f):
    return x-y-z-c-d-f
```

```python
minus(10,20,3,54,64,58)
```

```python
parameter :
    1.positional only before /
    2.keyword only after *
    3.standard -(p+k) 
```

```python
minus(10,20,3,54,d=64,f=58)
```
