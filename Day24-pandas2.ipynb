```python
# Converts the 'PassengerId' column to string (object)
df['PassengerId'] = df['PassengerId'].astype('str')
```

```python
# Shows the counts for each unique value in the 'Embarked' column
df['Embarked'].value_counts()
```

```python
# Convert 'Age' column to integer (int64)
df['Age'] = df['Age'].astype('int64')

# Convert 'Sex' column to the more memory-efficient 'category' type
df['Sex'] = df['Sex'].astype('category')

# Convert a numerical column to string
df['Ticket'] = df['Ticket'].astype(str)
```

```python
# Example: Convert 'Fare' which might contain placeholder strings
df['Fare'] = pd.to_numeric(df['Fare'], errors='coerce')

# Now, any non-convertible strings in 'Fare' are set to NaN.
# You can then use df.fillna() or df.drop_na() to handle them.
```

```python
# Convert a standard date string column
df['Date'] = pd.to_datetime(df['DateString'])
```

```python
Code,  Meaning,          Example
%Y,    Year with century,2025
%m,    Month as a number,12
%d,    Day of the month,  15
%b,    Abbreviated month name,Dec
```

```python
# Example for a date string like '15-12-2025'
df['PurchaseDate'] = pd.to_datetime(
    df['PurchaseDate'], 
    format='%d-%m-%Y'
)
```

```python
# Extracts the word after the comma and before the period
df['Title'] = df['Name'].str.extract('([A-Za-z]+)\.')
```

```python
Attribute,Purpose,Example
.dt.year,Extracts the year as an integer.,df['Date'].dt.year
.dt.month,Extracts the month (1-12).,df['Date'].dt.month
.dt.day,Extracts the day of the month.,df['Date'].dt.day
.dt.day_name(),"Extracts the weekday name (e.g., 'Monday').",df['Date'].dt.day_name()
.dt.is_leap_year,Returns a boolean (True/False).,df['Date'].dt.is_leap_year
```

```python
# Extracts the week of the year (01-52)
df['Week_of_Year'] = df['OrderDate'].dt.isocalendar().week

# Filters for data in the first quarter (Jan, Feb, Mar)
q1_data = df[df['OrderDate'].dt.quarter == 1]
```

```python
# Convert to 'category' type first
df['Severity'] = df['Severity'].astype('category')

# Define the logical order
new_order = ['Low', 'Medium', 'High', 'Critical']

# Apply the new order
df['Severity'] = df['Severity'].cat.reorder_categories(new_order, ordered=True)
```

```python
# Calculate the average Fare paid by passengers in each Pclass (1st, 2nd, 3rd)
df.groupby('Pclass')['Fare'].mean()
```

```python
# Group by 'Pclass' and calculate:
# 1. The min and max of 'Age'
# 2. The sum of 'Fare'
grouped_stats = df.groupby('Pclass').agg(
    {
        'Age': ['min', 'max'],
        'Fare': 'sum'
    }
)
```

```python
# Filters the DataFrame to show only rows where 'Age' is null
missing_age_rows = df[df['Age'].isnull()]
```

```python
# Fills all missing 'Cabin' values with the string 'Unknown'
df['Cabin'].fillna('Unknown', inplace=True)
```

```python
# Calculate the median of the 'Age' column
age_median = df['Age'].median()

# Fill the missing 'Age' values with the calculated median
df['Age'].fillna(age_median, inplace=True)
```
