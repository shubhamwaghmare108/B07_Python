```python
# Parent class
class Animal:
    def __init__(self, name):
        self.name = name

    def speak(self):
        raise NotImplementedError("Subclass must implement abstract method")

# Child class inheriting from Animal
class Cat(Animal):
    def __init__(self, name, breed):
        # Call the parent's constructor
        super().__init__(name)
        self.breed = breed

    # Override the speak method
    def speak(self):
        return f"{self.name} says meow!"

cat1 = Cat("Whiskers", "Siamese")
print(cat1.speak())
```

```python
class Parent1:
    def method1(self):
        print("Method from Parent1")
    def method2(self):
        print("Method from Parent1")

class Parent2:
    def method1(self):
        print("Method from Parent2")
    def method2(self):
        print("Method from Parent2")

class Child( Parent2,Parent1):
    def method3(self):
        print("Method from Child")

# Create an instance of the Child class
c = Child()
c.method1()
c.method2()
c.method3()
```

```python
# Parent class 1
class Vehicle:
    def __init__(self, brand, year):
        self.brand = brand
        self.year = year
        print("Vehicle initialized.")

    def display_info(self):
        print(f"Brand: {self.brand}, Year: {self.year}")

# Mixin class 1
class Motorized:
    def __init__(self, motor_type):
        self.motor_type = motor_type
        print("Motorized component added.")

    def run_motor(self):
        print(f"The {self.motor_type} motor is running.")

# Mixin class 2
class Amphibious:
    def __init__(self, propulsion_type):
        self.propulsion_type = propulsion_type
        print("Amphibious component added.")

    def float(self):
        print(f"The vehicle is floating using {self.propulsion_type}.")

# Child class using multiple inheritance
class AmphibiousCar(Vehicle, Motorized, Amphibious):
    def __init__(self, brand, year, motor_type, propulsion_type):
        # Call constructors of parent classes using super()
        super().__init__(brand, year)
        # We must explicitly call the __init__ of other mixins
        Motorized.__init__(self, motor_type)
        Amphibious.__init__(self, propulsion_type)
        print("AmphibiousCar created.")

    def drive(self):
        print("The vehicle is driving on land.")

# Create an object of the AmphibiousCar class
my_amphibious_car = AmphibiousCar("WaterCar", 2023, "Gasoline", "Jet Propulsion")

# Accessing methods from all parent and child classes
print("\n--- Displaying Info ---")
my_amphibious_car.display_info()
my_amphibious_car.run_motor()
my_amphibious_car.float()
my_amphibious_car.drive()
```
