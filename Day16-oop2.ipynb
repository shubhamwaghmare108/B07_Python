# OOP in python

```python
class Dog:
    # The __init__ method is a special method called a constructor.
    # It initializes the attributes of the object when it's created.
    def __init__(self, name, age):
        self.name = name  # 'self' refers to the object being created.
        self.age = age

    # This is a method. It's a function that belongs to the class.
    def bark(self):
        print(f"{self.name} says woof!")
    def walk(self):
        print(f"{self.name} says let's go for a walk")
```

```python
d1=Dog("Tommy",5) #(d1,"Tommy",5)
```

```python
d1.bark()
```

```python
d1.walk()
```

```python
class Person: # class name
    """
    adfshb
    """
    gender="Male"  # class variable-same for every object
    nationality="Indian"
    def __init__(self, name, age): # constructor function -run automatically when object is created.
        self.name = name # instance variable -different for all object
        self.age = age
    
p1 = Person("John",36) # object creation-class instance

print(p1.name) # attribute
print(p1.age)
print(p1.gender) # class variable
print(Person.gender) # class variable
print(p1.nationality)
print()

p2=Person("Priyanshu",21)

print(p2.name)
print(p2.age)
print(p2.gender)
print(p2.nationality)
print()
p3=Person("Shubham",29)

print(p3.name)
print(p3.age)
print(p3.gender)
print(p3.nationality)
print()
```
