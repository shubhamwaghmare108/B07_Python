```python
class Dog:
    # The __init__ method is a special method called a constructor.
    # It initializes the attributes of the object when it's created.
    def __init__(self, name, age):
        self.name = name  # 'self' refers to the object being created.
        self.age = age

    # This is a method. It's a function that belongs to the class.
    def bark(self):
        print(f"{self.name} says woof!")
    def walk(self):
        print(f"{self.name} says let's go for a walk")
```

```python
class Person: # class name
    """
    Want to create a basic or common indian person
    """
      # class variable-same for every object
    nationality = "Pakistani"
    def __init__(self, name,realName, age, gender): # constructor function -run automatically when object is created.
        self.name = name
        self.__name = realName # instance variable -different for all object
        self.age = age
        self.gender= gender
        self.__nationality = "Indian"
    
    def Identity(self,password=None):
        if password == 'Babarsher':
            return self.__name,self.__nationality

        elif password == None:
            return self.name,self.nationality
        else:
            return self.name,self.nationality
    
    def set_name(self,new):
        self.name = new
```
