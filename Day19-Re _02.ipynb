```python
import re

# Search for pattern (first occurrence)
re.search(pattern, string)        # Returns match object or None
re.match(pattern, string)         # Matches from start of string
re.fullmatch(pattern, string)     # Matches entire string

# Find all matches
re.findall(pattern, string)       # Returns list of matches
re.finditer(pattern, string)      # Returns iterator of match objects

# Replace and split
re.sub(pattern, replacement, string)     # Replace occurrences
re.subn(pattern, replacement, string)    # Replace with count
re.split(pattern, string)                # Split string by pattern
```
