```python
# List → To-Do Task Manager
```

```python
tasks = []

def add_task(task):
    tasks.append(task)
    print("Task Created Successfully")

def remove_task(task):
    if task in tasks:
        tasks.remove(task)
        print("Task Removed Successfully")
    else:
        print("Task not found")

def view_task():
    if len(tasks) == 0:
        print("Task list is empty")
    else:
        print("Task List".center(50, "*"))
        for i, j in enumerate(tasks, 1):
            print(i, "===>>", j)
```

```python
add_task("Learn Python")
```

```python
view_task()
```

```python
remove_task("Learn Python")
```

```python
view_task()
```
