```python
import numpy as np
import numpy.ma as ma

# Set seed for reproducibility
np.random.seed(42)
```

## 1. Array Creation – All Methods

### From Python sequences (list, tuple, set)

```python
# From list
a = np.array([1, 2, 3])
print("From list:", a)

# From tuple
b = np.array((4, 5, 6))
print("From tuple:", b)

# From set (order not guaranteed)
c = np.array({7, 8, 9})
print("From set:", c)
```

```python
# From any iterable (e.g., generator)
d = np.fromiter(range(5), dtype=int)
print("From iterable:", d)
```

### Ranges: `arange`, `linspace`, `logspace`, `geomspace`

```python
print("arange(5, 10):", np.arange(5, 10))
print("arange(0, 10, 2):", np.arange(0, 10, 2))
print("linspace(1, 10, 5):", np.linspace(1, 10, 5))
print("logspace(1, 3, 3):", np.logspace(1, 3, 3))   # 10^1, 10^2, 10^3
print("geomspace(1, 1000, 4):", np.geomspace(1, 1000, 4))
```

### Zeros, Ones, Empty, Full

```python
print("np.zeros(3):", np.zeros(3))
print("np.zeros((3,4)):\n", np.zeros((3,4)))
print("np.ones((2,3)):\n", np.ones((2,3)))
print("np.empty((2,2)) (uninitialised, may be garbage):\n", np.empty((2,2)))
print("np.full((2,2), 7):\n", np.full((2,2), 7))
```

### Identity / Diagonal matrices

```python
print("np.eye(3):\n", np.eye(3))
print("np.eye(3, k=1):\n", np.eye(3, k=1))
print("np.eye(3, k=-1):\n", np.eye(3, k=-1))
print("np.identity(3):\n", np.identity(3))
print("np.diag([1,2,3]):\n", np.diag([1,2,3]))
```

### Random arrays

```python
print("Uniform [0,1):", np.random.rand(3))
print("Standard normal:", np.random.randn(3))
print("Random integers 0-9:", np.random.randint(10, size=5))
print("Random integers 5-10, shape (2,3):\n", np.random.randint(5, 10, (2,3)))
print("Random sample (0,1):", np.random.random(4))
```

```python
# Random choice from an array
choices = np.array([10, 20, 30, 40])
print("Random choice:", np.random.choice(choices))
print("5 choices with replacement:", np.random.choice(choices, 5))

# Shuffle in-place
arr = np.array([1, 2, 3, 4, 5])
np.random.shuffle(arr)
print("After shuffle:", arr)

# Setting seed for reproducibility
np.random.seed(0)
print("Reproducible random:", np.random.rand(3))
```

### `*_like` functions

```python
base = np.array([[1, 2], [3, 4]])
print("np.zeros_like(base):\n", np.zeros_like(base))
print("np.ones_like(base):\n", np.ones_like(base))
print("np.full_like(base, 9):\n", np.full_like(base, 9))
```

### Grids: `meshgrid`, `mgrid`, `ogrid`

```python
x = np.array([1, 2, 3])
y = np.array([4, 5, 6])
X, Y = np.meshgrid(x, y)
print("meshgrid X:\n", X)
print("meshgrid Y:\n", Y)

print("mgrid[0:3,0:3]:\n", np.mgrid[0:3, 0:3])
print("ogrid[0:3,0:3]:\n", np.ogrid[0:3, 0:3])
```

### Minimum number of dimensions (`ndmin`)

```python
arr = np.array([1, 2], ndmin=5)
print("Shape with ndmin=5:", arr.shape)   # (1,1,1,1,2)
```

## 2. Array Attributes

```python
arr = np.array([[1, 2, 3], [4, 5, 6]])
print("shape:", arr.shape)
print("dtype:", arr.dtype)
print("size:", arr.size)
print("ndim:", arr.ndim)
print("itemsize (bytes per element):", arr.itemsize)
print("nbytes (total memory):", arr.nbytes)
```

## 3. Data Types and Conversion

```python
arr_int = np.array([1, 2, 3])
arr_float = arr_int.astype(float)
arr_str = arr_int.astype(str)
print("int → float:", arr_float, arr_float.dtype)
print("int → str:", arr_str, arr_str.dtype)

# Truncation when converting float to int
arr_float2 = np.array([1.9, 2.5, 3.1])
print("float → int (truncates):", arr_float2.astype(int))
```

## 4. Basic Element‑wise Operations

```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
print("a + b =", a + b)
print("a * b =", a * b)
print("a / b =", a / b)
print("a ** 2 =", a ** 2)
print("a > 2 =", a > 2)
```

## 5. Indexing & Slicing (1D, 2D, 3D)

```python
arr1 = np.arange(10)
print("arr1[2:7:2] =", arr1[2:7:2])
print("arr1[-3:] =", arr1[-3:])
```

```python
mat = np.arange(9).reshape(3, 3)
print("mat:\n", mat)
print("mat[1, 2] =", mat[1, 2])
print("mat[1, :] =", mat[1, :])
print("mat[:, 1] =", mat[:, 1])
print("mat[:2, 1:] =\n", mat[:2, 1:])
print("mat[::2, ::2] =\n", mat[::2, ::2])
```

```python
cube = np.arange(24).reshape(2, 3, 4)
print("cube[0, 1, 2] =", cube[0, 1, 2])
print("cube[1, :, :] (second block):\n", cube[1, :, :])
print("cube[..., 0] (last axis, first element):\n", cube[..., 0])
print("cube[0, 1, :] =", cube[0, 1, :])
print("cube[1, 1, 0:3:2] =", cube[1, 1, 0:3:2])
```

## 6. Boolean Indexing (Masking)

```python
arr = np.array([1, 2, 3, 4, 5])
mask = arr > 2
print("mask:", mask)
print("arr[mask]:", arr[mask])
print("arr[arr % 2 == 0]:", arr[arr % 2 == 0])

# Modify using mask
arr[arr > 3] = 99
print("After modification:", arr)
```

## 7. Fancy Indexing

```python
arr = np.array([10, 20, 30, 40, 50])
indices = [0, 2, 4]
print("arr[indices]:", arr[indices])

mat = np.arange(12).reshape(3, 4)
rows = [0, 2]
cols = [1, 3]
print("mat[rows, cols]:", mat[rows, cols])   # (0,1) and (2,3)

# Outer indexing (block)
print("mat[np.ix_(rows, cols)]:\n", mat[np.ix_(rows, cols)])
```

## 8. `np.where()` – Conditional Selection

```python
arr = np.array([1, 2, 3, 4, 5])
result = np.where(arr > 3, "high", "low")
print("where condition:", result)

# Get indices where condition is True
indices = np.where(arr > 3)
print("Indices of >3:", indices)
```

## 9. Set Operations

```python
a = np.array([1, 2, 3, 2, 4])
b = np.array([3, 4, 5, 6])
print("Unique a:", np.unique(a))
print("Intersection:", np.intersect1d(a, b))
print("Union:", np.union1d(a, b))
print("Difference (a - b):", np.setdiff1d(a, b))
print("Elements of a in b?", np.isin(a, b))
```

## 10. Logical Operations

```python
p = np.array([True, False, True])
q = np.array([False, False, True])
print("logical_and:", np.logical_and(p, q))
print("logical_or:", np.logical_or(p, q))
print("logical_not:", np.logical_not(p))
print("all(p):", np.all(p))
print("any(p):", np.any(p))
```

## 11. Vectorized String Operations (`np.char`)

```python
names = np.array(["alice", "BOB", "Charlie"])
print("Upper:", np.char.upper(names))
print("Lower:", np.char.lower(names))
print("Length:", np.char.str_len(names))
print("Replace 'a' with 'X':", np.char.replace(names, 'a', 'X'))

sentence = np.array(["one two three", "four five six"])
print("Split:", np.char.split(sentence))
```

## 12. Mathematical Functions (ufuncs)

```python
x = np.linspace(0, np.pi, 4)
print("x =", x)
print("sin(x) =", np.sin(x))
print("exp(x) =", np.exp(x))
print("log(x) =", np.log(x))
print("sqrt(x) =", np.sqrt(x))
```

## 13. Aggregations (sum, mean, min, max, cumsum, etc.)

```python
mat = np.array([[1, 2], [3, 4]])
print("Sum:", mat.sum())
print("Sum axis=0 (columns):", mat.sum(axis=0))
print("Sum axis=1 (rows):", mat.sum(axis=1))
print("Mean:", mat.mean())
print("Min:", mat.min())
print("Max:", mat.max())
print("Argmin (flattened):", mat.argmin())
print("Cumulative sum:", mat.cumsum())
print("Cumulative product:", mat.cumprod())
```

## 14. Statistical Functions

```python
data = np.array([10, 12, 23, 23, 16, 23, 21, 16])
print("Mean:", np.mean(data))
print("Median:", np.median(data))
print("Std:", np.std(data))
print("Variance:", np.var(data))
print("90th percentile:", np.percentile(data, 90))

# Mode via unique
vals, counts = np.unique(data, return_counts=True)
mode = vals[np.argmax(counts)]
print("Mode:", mode)

# Correlation
x = np.array([1, 2, 3, 4])
y = np.array([2, 4, 6, 8])
print("Correlation matrix:\n", np.corrcoef(x, y))

# Histogram
hist, edges = np.histogram(data, bins=5)
print("Histogram counts:", hist)
```

## 15. Sorting and Searching

```python
arr = np.array([3, 1, 2, 5, 4])
print("np.sort(arr):", np.sort(arr))        # copy
arr.sort()
print("arr.sort() in-place:", arr)

# argsort
arr2 = np.array([40, 10, 30, 20])
idx = np.argsort(arr2)
print("argsort indices:", idx)
print("arr2[idx]:", arr2[idx])

# searchsorted (array must be sorted)
sorted_arr = np.array([1, 3, 5, 7, 9])
print("Insert positions for [2,5,10]:", np.searchsorted(sorted_arr, [2, 5, 10]))

# lexsort (multiple keys)
last = np.array(['Smith', 'Jones', 'Smith', 'Adams'])
first = np.array(['John', 'Bob', 'Alice', 'Tom'])
ind = np.lexsort((first, last))
print("Sorted by last, then first:", list(zip(last[ind], first[ind])))

# partition (partial sort – first k smallest)
arr3 = np.array([7, 2, 5, 1, 9, 3])
print("Partition (k=3):", np.partition(arr3, 3))
```

## 16. Array Manipulation (reshape, resize, flatten, ravel, transpose)

```python
arr = np.arange(12)
print("reshape(3,4):\n", arr.reshape(3, 4))
print("reshape(2, -1):\n", arr.reshape(2, -1))

# resize (in-place, can change size)
a = np.array([1, 2, 3, 4])
a.resize(2, 3)   # fills with zeros if larger
print("a.resize(2,3):\n", a)

# np.resize (returns new array, repeats)
b = np.array([1, 2, 3])
c = np.resize(b, (2, 4))
print("np.resize (repeats):\n", c)

# flatten vs ravel
mat = np.array([[1, 2], [3, 4]])
print("flatten (copy):", mat.flatten())
print("ravel (view):", mat.ravel())

# transpose
print("mat.T:\n", mat.T)
print("transpose(1,0):\n", mat.transpose(1, 0))
```

## 17. Linear Algebra (`np.linalg`)

```python
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])

print("A @ B (matrix multiplication):\n", A @ B)
print("np.linalg.inv(A):\n", np.linalg.inv(A))
print("Determinant:", np.linalg.det(A))
print("Eigenvalues:", np.linalg.eigvals(A))

# Solve linear system
A2 = np.array([[3, 1], [1, 2]])
b = np.array([9, 8])
x = np.linalg.solve(A2, b)
print("Solution x:", x)
print("Check A2 @ x:", A2 @ x)

# SVD
U, s, Vt = np.linalg.svd(A2)
print("Singular values:", s)
```

## 18. Advanced Broadcasting (with `newaxis`, `broadcast_to`)

```python
# Using np.newaxis to add dimensions
row = np.array([1, 2, 3])[:, np.newaxis]   # shape (3,1)
col = np.array([4, 5, 6])[np.newaxis, :]   # shape (1,3)
print("row + col:\n", row + col)

# Explicit broadcasting
arr = np.arange(4).reshape(2, 2)
broadcasted = np.broadcast_to(arr, (3, 2, 2))
print("broadcast_to shape:", broadcasted.shape)
```

## 19. Structured Arrays (Record Arrays)

```python
dtype = [('name', 'U10'), ('age', 'i4'), ('weight', 'f8')]
data = np.array([('Alice', 25, 55.5), ('Bob', 30, 75.2)], dtype=dtype)
print("Structured array:\n", data)
print("Names:", data['name'])
print("Ages:", data['age'])

# Modify a field
data['age'][0] = 26
print("Modified ages:", data['age'])
```

## 20. Masked Arrays (`np.ma`) – Handle Missing Data

```python
x = ma.array([1, 2, -999, 4], mask=[0, 0, 1, 0])
print("Masked array:", x)
print("Mean ignoring mask:", x.mean())

# Mask values that are negative
y = np.array([1, -5, 3, -2])
y_masked = ma.masked_where(y < 0, y)
print("Masked negative values:", y_masked)
```

## 21. Fourier Transforms (`np.fft`)

```python
# 5 Hz sine wave
fs = 1000
t = np.linspace(0, 1, fs)
signal = np.sin(2 * np.pi * 5 * t)

# FFT
spectrum = np.fft.fft(signal)
freqs = np.fft.fftfreq(len(signal), 1/fs)
print("FFT shape:", spectrum.shape)
print("First 5 positive frequencies:", freqs[:5])

# Inverse FFT (reconstruction)
reconstructed = np.fft.ifft(spectrum)
error = np.max(np.abs(reconstructed - signal))
print("Reconstruction error:", error)
```

## 22. File I/O (Save & Load)

```python
# Binary .npy (single array)
arr = np.array([1, 2, 3])
np.save('my_array.npy', arr)
loaded = np.load('my_array.npy')
print(".npy load:", loaded)

# Multiple arrays in .npz
np.savez('multiple.npz', a=arr, b=np.array([4, 5, 6]))
data = np.load('multiple.npz')
print(".npz:", data['a'], data['b'])

# Text files
np.savetxt('data.txt', arr, delimiter=',')
txt_load = np.loadtxt('data.txt', delimiter=',')
print("Text load:", txt_load)

# Cleanup
import os
for f in ['my_array.npy', 'multiple.npz', 'data.txt']:
    if os.path.exists(f):
        os.remove(f)
```

## 23. Performance & Memory Efficiency

```python
# In-place operations (avoid copies)
arr = np.arange(1_000_000)
arr += 1   # faster than arr = arr + 1

# Using `out` parameter to reuse memory
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
result = np.empty(3)
np.add(a, b, out=result)
print("out result:", result)

# Views vs copies – views share memory
original = np.array([1, 2, 3, 4, 5])
view = original[1:4]          # view
copy = original[1:4].copy()   # explicit copy
view[0] = 999
print("Original after view modification:", original)  # changed

# memmap (for arrays larger than RAM) – commented
# fp = np.memmap('large.dat', dtype='float32', mode='w+', shape=(10000, 1000))
```

## 24. Axis Manipulation (moveaxis, expand_dims, squeeze)

```python
arr = np.arange(24).reshape(2, 3, 4)
print("Original shape:", arr.shape)

moved = np.moveaxis(arr, 0, 2)
print("After moveaxis(0→2):", moved.shape)

expanded = np.expand_dims(arr, axis=1)
print("After expand_dims(axis=1):", expanded.shape)

squeezable = np.array([[[1, 2, 3]]])   # shape (1,1,3)
print("Squeezed shape:", np.squeeze(squeezable).shape)
```

## 25. Polynomials (`np.polyfit`, `np.poly1d`)

```python
# Generate noisy quadratic data
x = np.linspace(0, 5, 50)
y = 2*x**2 - 3*x + 1 + np.random.randn(50)*5

# Fit degree 2 polynomial
coeffs = np.polyfit(x, y, 2)
print("Fitted coefficients (a, b, c):", coeffs)

# Create callable polynomial
p = np.poly1d(coeffs)
print("Polynomial:", p)
print("p(2):", p(2))
```

## 26. Advanced Indexing Tricks (Ellipsis, Combined Boolean/Fancy)

```python
# Ellipsis (...)
arr = np.arange(24).reshape(2, 3, 4)
print("arr[..., 0] (last axis first element):\n", arr[..., 0])
print("arr[0, ...] (first block):\n", arr[0, ...])

# Combine boolean mask with fancy indexing
mat = np.random.rand(5, 5)
row_mask = mat[:, 0] > 0.5
col_indices = [0, 2, 4]
result = mat[row_mask][:, col_indices]
print("Combined indexing shape:", result.shape)
```

## 27. Copy vs View – Detailed

```python
# .copy() creates a new independent array
original = np.array([1, 2, 3, 4])
copied = original.copy()
copied[0] = 999
print("Original after copy change:", original)   # unchanged

# .view() creates a new array that shares data
viewed = original.view()
viewed[0] = 888
print("Original after view change:", original)    # changed

# Slicing returns a view
slice_view = original[1:3]
slice_view[0] = 777
print("Original after slice view change:", original)

# Check memory sharing
print("original and copied share?", np.shares_memory(original, copied))
print("original and view share?", np.shares_memory(original, viewed))
```
