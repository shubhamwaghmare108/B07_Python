```python
b = 0
while True:
    
    try:
        10/b
        break
    except ZeroDivisionError:
        print("Division by zero is not possible")
        break
    except NameError:
        globals()['b'] = 10
    except Exception as e:
        print(e)
print(10+20)
```

```python
def calculate_age(year_of_birth):
    if year_of_birth > 2026:
        raise FutureBirthError("Future Birth not possible")
    else:
        return 2026 - year_of_birth

try:
    age = calculate_age(2030)
except FutureBirthError as e:
    print(f"Error: {e}")
```

```python
class FutureBirthError(Exception):
    pass
```
