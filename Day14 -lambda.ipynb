A lambda function is a small, anonymous, single-expression function. They are useful for short-term, throwaway functions.

```python
# Regular function
def square(x):
    return x * x

# Equivalent lambda function
square_lambda = lambda x: x * x

print(square_lambda(5)) # Output: 25
```

```python
square(5)
```

```python
d=(lambda x,y:x+y)(10,23)
```

```python
square = lambda x: x ** 2
print(square(4))  # Output: 16
```

```python
double = lambda x: x * 2
print(double(5))  # Output: 10
```

```python
lis=[1,2,3]
tuple(map(double,lis))
```

```python
nums = [1, 2, 3, 4, 5]

# Double each element
doubled = list(map(lambda x: x * 2, nums))
print(doubled)  # Output: [2, 4, 6, 8, 10]
```

```python
# Filter even numbers
evens = list(filter(lambda x: x % 2 == 0, nums))
print(evens)  # Output: [2, 4]
```

```python
def dou(lis):
    lis_res=[]
    for i in lis:
        lis_res.append(i*2)
    return lis_res
```
