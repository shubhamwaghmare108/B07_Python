A lambda function is a small, anonymous, single-expression function. They are useful for short-term, throwaway functions.

```python
# Regular function
def square(x):
    #print(x * x)
    return x*x

# Equivalent lambda function
square_lambda = lambda x: x * x
print(square(5))
print(square_lambda(5))
```

```python
nums = [1, 2, 3, 4, 5]

# Double each element
doubled = list(map(lambda x: x * 2, nums))
print(doubled)  # Output: [2, 4, 6, 8, 10]
```

```python
# Filter even numbers
evens = list(filter(lambda x: x % 2 == 0, nums))
print(evens) 
list(filter(lambda x:x>2,nums))
```

```python
x = 60

def add():
    #nonlocal x
    #global x
    x = 40
    print(x)
    print(locals())
    
    def add2():
        #nonlocal x
        x = 20
        print(x)
        print(locals())

        def add3():
            nonlocal x
            x += 30
            print(x)
            print(locals())
        
        add3()
        print(locals())
    
    add2()

print(x)
add()
print(x)
```
