```python
def factorial(n):
    if n == 0 or n == 1:   # base case
        return 1
    return n*factorial(n-1)

print(factorial(5))
```

```python
y=10
def sdfs():
    global y
    y=20
    print(y)
print(y)
sdfs()
print(y)
```

```python
y=10
def enclosing():
    global y
    y=20
    print(y)
    def local_():
        nonlocal y
        y+=30
        print(y)
    local_()
    print(y)
print(y)
enclosing()
```
